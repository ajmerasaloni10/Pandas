# How to remove the index name. 
In this case ' country ' 

```python
drinks.index.name = None
drinks.head()

# index name removed.
```

# How to move index back to data frame

1. assign back the index name 
2. reset_index 

```python
drinks.index.name = 'country'
drinks.reset_index(inplace=True)
drinks.head()
```
